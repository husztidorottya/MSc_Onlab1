```python
import sys
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import nltk

# set plot properties
matplotlib.style.use('ggplot')
matplotlib.pyplot.rcParams['figure.figsize'] = (25, 13)
matplotlib.pyplot.rcParams['font.family'] = 'sans-serif'
matplotlib.pyplot.rcParams['font.size'] = 20

# read the experiment result
df = pd.read_csv('log_parameters.csv')

# visualize the score according to the average of epoch
df.groupby(by=[df.epoch]).mean().plot(y='accuracy', color='blue')

# show table grouped by epoch
columns = ['accuracy']
df_epoch = df[columns]
df_epoch.groupby(by=[df.epoch]).mean()
```
