```python
import sys
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import nltk

# set plot properties
matplotlib.style.use('ggplot')
matplotlib.pyplot.rcParams['figure.figsize'] = (25, 13)
matplotlib.pyplot.rcParams['font.family'] = 'sans-serif'
matplotlib.pyplot.rcParams['font.size'] = 20

# read the experiment result
df = pd.read_csv('log_train.csv')

# visualize the score according to the average of epoch
df.groupby(by=[df.epoch]).mean().plot(y='val_acc', color='blue')

# show table grouped by epoch
columns = ['val_acc']
df_acc_per_epoch = df[columns]
df_acc_per_epoch.groupby(by=[df.epoch]).mean()
```

```python
# visualize the score according to the average of epoch 
df.groupby(by=[df.epoch]).mean().plot(y='val_loss')

# show table grouped by epoch 
columns = ['val_loss']
df_loss_per_epoch = df[columns]
df_loss_per_epoch.groupby(by=[df.epoch]).mean()
```
